## 데이터 베이스
![](https://img1.daumcdn.net/thumb/R720x0.q80/?scode=mtistory2&fname=http%3A%2F%2Fcfile26.uf.tistory.com%2Fimage%2F995ABD335A087C94345B5E)
### 특징
- 대용량 데이터를 효율적으로 편리적하게 관리하는 시스템 
- 응용 시스템 들이 공유해서 사용하는 운영 데이터들이 구조적으로 통합된 모임
- 데이터베이스 구조는 **데이터 모델**에 의해 결정된다.
    - 관계형
    - 계층형
    - 망형
- 모든 데이터가 **중복을 최소화**
- 한 attribute마다 제약조건을 달아 무결성을 유지한다. ex) INTEGER != TEXT
- 데이터베이스는 운영 데이터 뿐만 아니라 그 데이터에 관한 설명(**scheme : 데이터베이스 구조** 또는 **metadata: 데이터 설명**)까지 포함
- 프로그램(op_code)과 데이터(operand) 간의 독립성 유지-> 프로그램에 영향을 주지 않으면서 데이터베이스 구조를 변경할 수 있음.
- 효율적으로 접근이 가능하고 질의를 할 수 있음

- **시스템 카탈로그(데이터 사전:scheme:논리적)** 와 **데이터베이스(데이터 파일:block:물리적)**로 구분
- 사용자의 질의를 빠르게 수행할 수 있는 인덱스 등의 접근 경롤를 DBMS가 자동적으로 선택하여 수행.
- 동시성 접근및 동시성 제어
    - DBMS는 다중 프로그래밍의 개념을 이용하기 때문에 한 트랜잭션을 실행하는 중에 다른 트랜잭션이 끼어들어 실행할 수 있다.
    - transition은 많은 질의문을 묶어 하나의 작업을 수행하는 단위로서, 하나의 thread이다.
    - transition은 thread이고 dbms는 다중 프로그래밍이여서 동시에 작업이 가능하다. 
    - 하지만 그로인해 데이터의 무결성이 깨지게 되고 신뢰성이 없는 데이터들로 꽉 차게 될 수 있다.

### SQL

> #### DDL
- 데이터 정의어
- 데이터베이스를 생성하고 데
- : CREATE, ALTER, ..

> #### DML
- 데이터 조작어
- 데이터를 효율적으로 질의(query, SELECT)하고 수정(UPDATE, DELETE, INSERT)할 수 있음
- : SELECT, UPDATE, INSERT, DELETE

> #### DCL
- 데이터 제어어
- 시스템의 고장이나 권한이 없는 사용자부터 데이터를 안전하게 보호할 수 있음.
- 백업과 회복기능

### scheme
- 전체적인 데이터베이스 구조를 나타낸다(논리적).
- **데이터 사전에 저장되어 있어 데이터 파일을 처리할 때 데이터가 있는 장소를 찾으려고 할때 사용**.

### DBMS
- DataBaseManagementSystem: 데이터 베이스 관리 시스템
- 데이터베이스를 정의(DDL)하고 질의어(DML)를 지원하고, 리포트를 생성(DCL)하는 등의 작업을 수행하는 **소프트웨어**
- DBMS에서 원하는 정보를 찾기 위해서는 DBMS를 주기억장치에 적재시키고 DBMS를 통해 디스크의 블록들을 주기억 장치로 읽어들여야 한다.
- ex)oracle, mysql, sqlite, ... 

### ANSI/SPARC

#### 외부스키마
- 각 사용자의 뷰(가상 테이블, 서브스키마) -> 실존x, 한개 또는 여러 테이블로 부터 원하는 튜플을 가져와 하나의 view로 만드는 것
- 개념 스키마로 부터 다수의 서로 다른 뷰가 제공
- 뷰의 사용목적: 원하는 테이블을 추출하기 위해 일일이 sql을 치지 않고 view을 따로 만들어 저장하고, 필요할 때마다 불어오는 방식(일종의 함수).

#### 개념스키마
- 전체 조직체(테이블)의 전반적인 정보(스키마), 실존 
- 물리적 구현x, 논리적 표현(어떤 데이터(타입, 이름)이 저장되어 있는지)
- DB에는 한개의 개념 스키마만 존재
- 데이터베이스에 대한 사용자 공동체의 뷰를 나타냄
- 개념스키마가 바꾸어도 외부스키마에 영향x(데이터독립성)
#### 내부스키마
- 실제, 물리적 데이터 구조에 대한 스키마
- DBMS 지시로 운영체제가 관리함
- 내부 구조가 바뀌어도 개념스키마에게 영향x(데이터독립성)

### 데이터 모델

#### Hierarchical DBMS(계층)
- 레코와 링크로 구성된 트리 형태의 구조로 조직된 것을 뜻함.
- 반복적인 부모-자식 관계 정보를 표현
- 오늘날의 데이터베이스들에서 거의 쓰이지 않지만 정보를 기록하는 기법은 다른 곳에서 쓰이고 있다.(파일 시스템, xml문서,...)

#### Network DBMS(네트워크)
- 그래프 구조를 기반으로 한다.
- 레코드를 노드로 나타내고, 레코드와 레코드의 관계는 간선으로 나타낸다. 
- 설계가 복잡하고 데이터 종속성 문제가 있어 구조를 변경하기 어려워 확장성이 떨어진다.

### Relational DBMS
- 행과 열로 구성된 2차원 테이블에 데이터(레코드)를 저장한다. 
- 테이블은 고유한 이름을 가지며, 데이터베이스에서 표현하는 하나의 엔티티에 관한 정보를 저장.
- 노드(레코드)간의 링크 또는 간선등이 없어 구조를 변경하기 쉽다.
- 계층DBMS나 네트워크DBMS에 비해 개념이 단순하여 현재까지도 가장 널리 사용된다.

#### OODBMS(Object-Oriented-DBMS)
- 정보를 객체의 형태로 표현하는 데이터베이스 모델.
- 객체 지향 프로그래밍의 패러다임을 기반으로 한다. 



### 관계 데이터 모델
![](https://t1.daumcdn.net/cfile/tistory/99269B375B3530C51A)
- 릴레이션: 행과 열로 구성된 2차원 테이블(릴레이션), 엔티티에 관한 데이터를 저장.
- 엔티티: 데이터의 집합으로 데이타베이스에 표현하려고 하는 유형, 무형의 객체로서 서로 구별되는 것
- 어트리뷰트: 릴레이션의 각 열, 속성(필드)
- 튜플: 릴레이션의 각행으로, 튜플(레코트), 어트리뷰트의 집합
- 차수: 튜플을 구성하는 속성의 개수
- 카디널리티: 튜플의 개수
- 스키마: 데이터베이스의 구조와 제약 조건에 관한 전반적인 명세를 기술한 메타데이터의 집합, 내포라고도 한다.
- 도메인: 하나의 어트리뷰트(속성)이 가질 수 있는 값들의 범위및 집합
- 키: 각 튜플을 유일하게 구별할 수 있는 하나 이상의 속성 모임.

### 릴레이션 스키마
- 전체적인 데이터베이스 구조를 뜻하며 자주 변경되지는 않는다.
- 데이터베이스의 구조와 제약 조건에 관한 전반적인 명세를 기술한 메타데이터의 집합, 내포라고도 한다.
- 릴레이션에 어떤 정보가 담길지를 정의한다

- ex) 저자(author)와 도서이름(title)이 적힌 책을 가진 BookTable이라는 새로운 도서관을 만들어라 -> CREATE TABLE BookTable (author,title) 
- **CREATE TABLE 테이블이름 (어트리뷰트명 도메인, ..)** == **CREATE SCHEME 스키마이름 (스키마 요소, ..)**
    - 해당 명령어는 DDL(데이터 정의어)이다.
    - 다음 명령어는 TABLE 또는 스키마를 생성하는 명령어이다.
    - 스키마는 데이터베이스의 구조와 제약 조건에 관한 전반적인 명세를 기술하는 것이다. 다음 명령을 보면 전반적인 해당 테이블의 내용을 알수 있다.
    - 릴레이션에 어떤 정보가 담길지를 정의한다.
    
### 릴레이션 인스턴스
- 릴레이션 스키마에 실제로 저장된 데이터의 집합
- 튜플들의 집합이며 외연(extension)이라고도 한다.
- 삽입, 삭제, 수정시 리레이션 인스턴스는 변한다.
- 시간의 흐름에 따라 변한다.

- **명령어**
    - **SELECT 속성,.. FROM 테이블이름 WHERE 속성='해당속성'**
        - ex) 도서관(bookTable)에서 자료구조 도서(title)의 저자(author) 이름을 출력하라 -> SELECT author FROM BookTable WHERE title="자료구조";
    - **INSERT**
        - ex) 저자(author)가 홍길동인 자료구조 도서(title)를 도서관(BookTable)에 새로 추가 해라 -> INSERT INTO BookTable(author, title) VALUES ("홍길동","자료구조");
    - **UPDATE**
        - ex) 도서관(BookTable)에서 저자가 자료구조 도서(title)의 저자(author)를 김길동으로 바꾸세요 -> UPDATE BookTable SET author="김길동" WHERE title="자료구조"
    - **DELETE**
        - ex) 도서관(BookTable)에서 저자가 자료구조 도서(title)를 폐기 해주세요 DELETE FROM BookTable WHERE title="자료구조"

### 릴레이션의 특징
- 릴레이션을 구성하는 튜플을 유일하게 식별하기 위해 속성들의 부분빕합을 키로 설정.
- 속성의 값은 논리적으로 더 이상 쪼갤수 없는 원자 값만을 저장.
- 튜플들의 삽입, 삭제등의 작업드로 인해 릴레이션은 시간에 따라 변함.
- 한 애트리뷰트(속성)은 하나의 도메인 값만 가진다.
- 한 릴레이션에는 키를 가지므로 똑같은 튜플이 포함될수 없다. 따라서 릴레이션에 포함된 튜플들은 모두 상이하다.
- 한 릴레이션에 포함된 튜플 사이에는 순서가 없다.

### 릴레이션의 키
- 릴레이션의 키는 각 튜플을 고유하게 식별할 수 있는 하나 이상의 애트리뷰트들의 모임.
- 일반적으로 키는 두 리레이션을 서로 연관시키는데 사용된다. 
- 리레이션의 튜플을 접근하는 속도를 높이기 위해 키에 인덱스를 만드는데, 키가 작을수록 인덱스의 크기가 줄어들고 인덱스를 검사하는 시간이 단축된다.
- 종류
    - candidate key(후보키)
        - 릴레이션을 구성하는 속성들중에서 튜플을 유일하게 식별할 수 있는 속성들의 부분집합을 의미
        - 모든 릴레이션은 반드시 하나 이상의 후보키를 가져야한다.
    - primary key(기본키)
        - 후보키중에서 선택한 주키 이므로 테이블에서 기본키는 하나만 만들 수 있음.
        - 한 릴레이션에서 특정 튜플을 유일하게 구별할 수 있는 속성
        - Null 값을 가질수 없다.
        - 기본키로 정의된 속성에는 동일한 값이 중복되어 저장될 수 없다.
        - 자연스러운 기본키를 찾을 수 없는 경우에 레코드 번화와 같이 종종 인위적인 키 애트리뷰트를 릴레이션에 추가할 수 있는데 이 키를 대리키라고 한다.
    - alternate key(대체키)
        - 후보키가 둘 이상일때 기본키를 제외한 나머지 후보키들을 말한다.
        - 보조키라고도 한다.
    - super key(슈퍼키)
        - 슈퍼키는 한 릴레이션 내에 있는 속성들의 집합으로 구성된 키
        - 릴레이션을 구성하는 모든 튜플에 대해 유일성은 만족하지만, 최소성은 만족시키지 못한다. 
    - foreign key(왜래키)
        - 관계(Relation)를 맺고 있는 릴레이션 R1, R2에서 릴레이션 R1이 참조하고 있는 릴레이션 R2의 기본키와 같은 R1 릴레이션의 속성
        - 외래키는 참조되는 리레이션의 기본키와 대응되어 릴레이션 간에 참조관계를 표현하는 데 중요한 도구로 사용.
        - 왜래키로 지정되면 참조 테이블의 기본키에 없는 깞은 입력할 수 없음
        
### 외래키의 유형
#### 다른 릴레이션의 기본키를 참조하는 외래키(연결)
- 예시 고객 릴레이션과 주문 릴레이션에서 고객 릴레이션의 외래키는 고객아이디고 그에 해당하는 기본키는 주문 릴레이션의 주문고객이다. 고객아이디 값의 범위는 주문고잭 값 범위(도메인)안에 있어야 된다. 

#### 자체 릴레이션의 기본키를 참조하는 외래키
- 고객 릴레이션에서 곡개아이디가 기본키이고 이에 해당하는 외래키는 추천고객이다. 이처럼 고객 릴레이션은 기본키와 외래키가 하나의 릴레이션에서 표현되며, 이것은 고객릴레이션이 자기 자신의 릴레이션과 관계를 맺고 있음을 의미한다. 
- 정지영 고객은 다른 고개의 추천고객을 기입하지 않고 가입했기 때문에 추천고객이 null으로 되어있다. 이처럼 외랙키는 기본키를 참조하지만 기본키가 아니기 때문에 널값을 가질 수 있다.

#### 기본키의 구성요소가 되는 외래키
상담 릴레이션의 학번 속성은 학생 릴레이션의 기본키인 학번 속성을 참조하는 외래키고, 담당교사 속성은 교사 리레이션의 기본키인 교사번호 속성을 참조하는 외래키다. 이처럼  하나의 릴레이션에는 외래키가 여러개 존재할수도 있다. 이 두개 왜래키를 합쳐서 기본키가 된다

### 데이터 무결성과 제약조건

#### 데이터 무결성
- 데이터의 무결성은 데이터의 정확성, 일관성, 유효성이 유지되는 것을 말한다.
- 데이터의 무결성을 유지하는 것은 데이터베이스 관리시스템의 중요한 기능이며 주로 데이터에 적용되는 연산에 제한을 두어 데이터의 무결성을 유지한다.

#### 데이터 무결성 제약조건
- 데이터베이스의 일관성을 보장하기 위해 일관된 데이터베이슬 상태를 정의하는 규칙들을 묵시적으로 또는 명시적으로 정의하는 것
- 종류
    - 도메인 무결성 제약조건: 튜플 안의 모든 속성이 각각의 도메인 집합 안에 존재하는 값을 취해야 된다는 것.
    - 키무결성 제약조건:키 애트리뷰트에 중복된 값이 존재해서는 안된다는 것
    - 엔티티 무결성 제약조건: 릴레이션의 기본 키를 구성하는 애트리뷰트는 널 값을 가질 수 없다는 것
    - 참조 무결성 제약조건(외래 키 무결성 제약조건): 두 릴레이션의 연관된 투플들 사이의 일관성을 유지하는데 사용

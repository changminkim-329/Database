## DataBase란?
- 여러 사람에 의해 공유되어 사용될 목적으로 통합하여 관리되는 데이터의 집합
- 자료항목의 중복을 없애고 자료를 구조화하여 저장함으로써 자료 검색과 갱신의 효율을 높인다.
- 방대한 데이터를 효율적으로 관리하기 위해 컴퓨터에 통합.저장한 것으로, **데이터베이스관리시스템(DBMS)**이라는 프로그램을 이용하여 관리한다.
![](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=http%3A%2F%2Fcfile9.uf.tistory.com%2Fimage%2F993845445A67253915C638)
- 릴레이션: 행과 열로 구성된 2차원 테이블(릴레이션), 엔티티에 관한 데이털를 저장.
- 튜플: 릴레이션의 각행, 튜플(레코드 or 엔트리), 필드의 집합.
- 속성: 릴레이션의 각 열, 속성(필드).
- 도메인: 속성 하나가 가질 수 있는 값의 집합, "성별"속성의 도메인: 2{'남자','여자'}
- 차수: 튜플을 구성하는 속성의 개수, 위의 그림의 차수: 5
- 카디널리티: 튜플의 개수
- 키: 각 튜플을 유일하게 구별할 수 있는 하나 이상의 속성 모임.
    - **후보키**: 각 튜플을 유일하게 구별할 수 있는 '최소한'의 속성. ex) 학번 또는 (학과명, 이름): 하나라도 빠지면 구별 X
    - **기본키**: 후보키가 두 개 이상이면 대표로 삼을 키. PRIMARY KEY
    - **외래키**: 다른 릴레이션의 기본키를 참조하며, 릴레이션과 릴레이션 상이의 관계를 나타내는 키.

### 데이터 베이스의 특징
- 통합되어 저장: 여러 가지 정보(속성)을 묶어 의미있는 정보를 제공(레코드).  
- 물리적인 기억장치에 보관
- 권한이 있으면 데이터 수정이 가능
- 여러 사람이 사용 가능및 동시 공유가 가능
    - 여러 사용자와 응용 프로그램이 동시에 접근하여 데이터 검색
- 내용에 따라 참조한다(key)

### 데이터베이스 관리 시스템(DBMS)
- 사용자와 응용 프로그램에게 편리하고 효율적인 데이터베이스 사용 환경을 제공하는 소프트웨어.

- 특징
    - 데이터베이스 구조를 명시.

    - 데이터를 효율적으로 검색및 수정.
    - **데이터 독립성이 가능**
        - Ex) Java Virtual Machine

### 스키마(Schema)
- 데이터베이스의 구조와 제약조건에 관해 전반적인 명세를 기술한 것, 메타 데이터(목적을 위한 데이터) 이다.
    - 개체의 특성을 나타내는 **속성**과 속성들의 집합으로 이루어진 **개체(Entity)**, 개체 사이에 존재하는 **관계**에 대해 정의.
    - 속성, 개체 그리고 관계에서 유지해야 할 제약조건을 기술.
- DB내에서 어떤 구조로 데이터가 저장되는 가를 나타내는 데이터베이스 구조.
- 데이터 사전에 저장.

- 종류
    - **외부 스키마(사용자 뷰)**: 실제 존재하는 데이터들을 어떤 형식, 구조, 배치 화면을 통해 사용자에게 보여주는 것(브라우저)
    - **개념 스키마(전체적인 뷰)**: 데이터 베이스의 전체적인 논리적 구조로서, 모든 응용 프로그램이나 사용자들이 필요로 하는 데이터를 종합한 조직 전체의 데이터 베이스로 하나만 존재한다.
    - **내부 스키마**: 개념 스키마를 디스크 기억장치에 물리적으로 구현하기 위한 방법을 기술한 것

### 자료형
- INTEGER : 정수 유형
- TEXT : 문자열 유형
- REAL : 부동소수점수(floating-point)유형
- NUMERIC : 전체 자릿수와 소수 자릿수가 고정된 숫자 데이터 형식, 불린, 날짜, 날짜-시간
- BLOB : 이진 대형 객체(Binary Large Object)- 이진 데이터의 모임
    -  일반적으로 그림, 오디오, 또는 기타 멀티미디어 오브젝트

### 데이터베이스 언어

#### 데이터 정의어(DDL)
- 데이터베이스를 구축하거나 수정할 때 사용하는 언어.
- 테이블을 생성하고, 생성된 테이블을 수정, 삭제하는 명령어.
    - CREATE TABLE 데이터베이스 이름: 테이블을 생성한다.
        - **CREATE TABLE 릴레이션명**
            - ex) CREATE TABLE DRINK
        - **(속성1 자료형1 [NULL | NOT NULL], 속성2 자료형2 [NULL | NOT NULL], ..., 속성n 자료형n [NULL | NOT NULL]);**
            - (_id INTEGER PRIMARY KEY AUTOINCREMENT, NAME TEXT, DESCRIPTION TEXT);
    - ALTER TABLE 데이터베이스 이름: 테이블을 수정(속성 추가, 변경, 삭제)한다.
        - **ALTER TABLE 릴레이션명 ADD 속성명 자료형;**
            - ex) ALTER TABLE DRINK ADD COLUMN FAVORITE NUMERIC
        - **ALTER TABLE 릴레이셩명 MODITY 속성명 자료형;**
        - **ALTER TABLE 릴레이션명 DROP 속성명;**
    - DROP TABLE 데이터베이스 이름: 테이블을 삭제한다.
        - **DROP TABLE 릴레이션명**
    - TRUNCATE TABLE 데이터베이스 이름: 테이블의 데이터를 삭제한다. TRUNCATE TABLE 명령어로 삭제된 행들은 복구할 수 없다.
    - RENAME TABLE 오래된 데이터베이스 이름 TO 새로운 데이터베이스 이름: 테이블 이름을 변경한다.
        - **RENAME TABLE 릴레이션**
    
#### 데이터 조작어(DML)
- 테이블에 데이터를 추가하거나 데이터를 수정, 삭제, 검색(Query)할 때 사용하는 언어.
- 사용자와 DBMS 사이의 인터페이스를 제공.
- 질의어를 통해 구현. Ex) https://www.naver.com
- 데이터 조작 명령어
    - INSERT : 데이터를 추가한다.
        - **INSERT INTO 릴레이션 릴레이션명(속성1, 속성2, ...,속성n)** ex) INSERT INFO DRINK(name, description) 
        - **VALUES (값1, 값2, ..., 값n)** ex) VALUES ("Latte","Tasty")
            - name, description 속성에 "Latte", "Tasty", 나머지 속성은 null로 레코드를 추가
    - UPDATE : 데이터를 수정한다.
        - **UPDATE 릴레이션명** ex) UPDATE DRINK(DRINK라는 테이블을 업데이트)
        - **SET 속성1=수식1, 속성2=수식2, ..., 속성n=수식n** ex) SET DESCRIPTION = "Tasty"( DESCRIPTION 속성 값을 "Tasty"로 바꾸기 )
        - **WHERE 조건;** ex) WHERE NAME="Latte"("Latte"라는 속성를 가진 레코드를 선택) 
    - DELETE : 데이터를 삭제한다.
        - **DELETE FROM 릴레이션명** ex) DELETE FROM DRINK
        - **WHERE 조건** ex) WHERE NAME="Latte"("Latte"라는 속성을 가진 레코드를 삭제)
    - SELECT : 데이터를 검색한다.
        - **SELECT [DISTINCT] 속성_리스트** : 걺색 결과에 포함시킬 속성의 목록을 나열.
        - **FROM 릴레이션_리스트** : 검색 대상이 되는 릴레이션을 나열.
        - **[WHERE 조건]** : 결과 리레이션의 튜플이 만족해야 하는 조건을 기술.
        - **[GROUP BY 속성_리스트]** : 결과 릴레이션의 튜플을 그룹하할 때 GROUP BY 절에서 명시한 속성 값과 동일한 튜플을 그룹으로 묶는다.
        - **[HAVING 조건]** : GROUP BY 로 그룹을 묶을 때 만족해야 하는 조건을 기술, HAVING 절과 GROUP BY 절은 언제나 함께 사용.
        - **[ORDER BY 속성_리스트[ASC | DESC]];** : 검색 결과의 정렬 방법(오름차순 or 내림차순)을 지정.
    
#### 데이터 제어어
- 데이터를 보호하고 관리할 때 사용하는 언어로, 데이터베이스 사용자에게 부여된 권한을 정의하는 언어.
- 데이터 접근 권한을 다루는 역할을 하는 명령어
    - Privileges - 객체에 대한 권한 부여 등을 제어.
    - GRANT : 권한을 준다.
    - REVOKE : 권한을 제거한다.

### Transaction
- 논리적인 일의 단위. 
- 하나의 묶음 처리가 되는 것.
- 양측이 다 원하는 결과물을 얻어야만 정상적으로 처리되는 것.

### Commit
- DML언어를 사용할 때는 항상 COMMIT도 함께 해줘야 한다.
- DML언어로 작업한 모든 것을 정상적으로 처리하겠다고 확정하는 명령어
- 트랜젝션의 처리 과정을 데이터베이스에 반영하기 위해서, 변경된 내용을 모두 영구 저장한다.
- COMMIT 수행하면, 하나의 트랜젝션 과정을 종료하게 된다.
- TRANSACTION(INSERT, UPDATE, DELETE)작업 내용을 실제 DB에 저장한다.
- **Auto Commit** 자동으로 commit을 해주는 설정.

### RollBack
- 작업 중 문제가 발생했을 때, 트랜젝션의 처리 과정에서 발생한 변경 사항을 취소하고, 트랜젝션 과정을 종료시키는 것.
- 트랜젝션 이전의 상태로 되돌린다.
- Rollback : 다시 되돌리는 능력.
